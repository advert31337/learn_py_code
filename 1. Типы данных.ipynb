```python
# посмотрим на возникающие неточности в компьютерных расчетах
a = 0.1 + 0.3 + 0.2
print(a, '<- обратите внимание на неточность мы не получили значение 0.6')
```

```python
# пробуем получить значение заданной точности
from decimal import Decimal

sum1 = Decimal("0.1") +  Decimal("0.3") +  Decimal("0.2")
print(sum1, '<- обратите внимание на неточность мы получили значение 0.6')   

# можно получить заданную точность и другим способом
sum2 = Decimal(0.1 + 0.3 + 0.2).quantize(Decimal('.00'))
print(sum2, '<- обратите внимание на неточность заданную методом quantize()')
```

Немного практики

```python
from datetime import datetime, timedelta

start_date = '2019-01-01'
end_date = '2019-01-07'

start_date_dt = datetime.strptime( start_date, '%Y-%m-%d')
end_date_dt = datetime.strptime( end_date, '%Y-%m-%d')

current_day = start_date_dt

while current_day <= end_date_dt:
    print(current_day)
    current_day += timedelta(days = 1)
```

```python
from datetime import datetime, timedelta


def date_range(start_date, end_date):
    date_range_list = []
   # current_date    = start_date
    current_date_dt = datetime.strptime(start_date, '%Y-%m-%d')
    end_date_dt     = datetime.strptime(end_date, '%Y-%m-%d')
    while current_date_dt <= end_date_dt:
        date_range_list.append(current_date_dt.strftime('%Y-%m-%d'))
        current_date_dt += timedelta(days = 1)
        #current_date = current_date_dt.strftime('%Y-%m-%d')
    return date_range_list 


start_date = '2019-01-01'
end_date = '2019-01-07'
x = date_range(start_date, end_date)
print(x)
```

```python
from datetime import datetime,

def get_dttstmp(post_time):
    lst = []
    for t in post_time:
        time_string = datetime.now().strftime('%d.%m.%Y') + ' ' + t
        lst.append(datetime.strptime(time_string, '%d.%m.%Y %H:%M').timestamp())
    return lst
    

post_time = ['14:13', '16:32', '20:34', '21:23', '22:12', '23:59']
print(get_dttstmp(post_time))
```

### **Присваивание**

```python
# стандартное присваивание
a = 10
print(a, '<- стандартное присваивание')

# присваивание "распаковкой"*"
list1 = (1, 99, 777)
a, b, c = list1
print(a, b, c, '<- присваивание "распаковкой"')

# переопределние
a = 10
b = -99
a, b = b, a
print(a, b, '<- присваиваине переопределнием')

# множественное присваивание
a = b = c = 0
print(a, b, c, '<- множественное присваивание')

# комбинированное присваивание
a = 0
a += 1
print(a, '<- комбинированное присваивание')
```

#### **Имена переменных**
Для работы с объектами для начала вам нужно связать имя переменной с объектом в памяти. Делается это очень просто командой:  
>a = 5

Имена переменных могут состоять букв цифр и прочих символов, но первый символ должен быть обязательно буквы или нижнего подчеркивания '\_' 
Python 3 поддерживает работу с кириллическими символами, но все же лучше использовать латинские символы.  
При написании всех объектов в Python **учитывается регистр** символов!  
Нельзя использовать зарегистрированные слова, рассмотренные ранее.  
По возможности все названия а также описание функций, комментарии писать на английском языке.  
В языке Python есть определенные соглашения при написании кода, которых придерживаются все программисты.  
Вот некоторые из них:  
- Константы записываются большими буквами CONSTANT
- Функции и методы пишутся в нижнем регистре в snake-стиле my_new_functions  
- Классы записываются в camel-стиле MyNewClass  
- В качестве отступа используются 4 пробела  

В сообществе Python существует целый документ с рекомендациями по написанию кода. Это **PEP 8** (Python Enhancement Proposals), который называется **"Style Guide for Python Code"** (Стиль кода в языке Python).  
Всем программистам на Python настоятельно рекомендуется ознакомиться с документом и придерживаться рекомендаций.
